```python
import camelot
```

```python
import pandas as pd
```

```python
tables = camelot.read_pdf("monthlyannualtemp.pdf", flavor="stream", pages="1-4")
tables
```

```python
tables[2].df
```

```python
df = pd.concat([
    tables[0].df,
    tables[1].df,
    tables[2].df,
    tables[3].df,
], ignore_index=True)
df
```

```python
#Change the columns
df = df.rename(columns={
    0: 'YEAR',
    1: 'JAN',
    2: 'FEB',
    3: 'MAR',
    4: 'APR',
    5: 'MAY',
    6: 'JUN',
    7: 'JUL',
    8: 'AUG',
    9: 'SEP',
    10: 'OCT',
    11: 'NOV',
    12: 'DEC',
    13: 'ANNUAL'
})
df = df[df['YEAR'] != 'YEAR']
df
```

```python
#drop any rows that have unnecessary values in the rebounds columns
df = df[df.MAR != "Average Monthly & Annual Temperatures at Central Park"]
```

```python
df = df[df.JUN != "Last Updated: 5/9/21"]
```
